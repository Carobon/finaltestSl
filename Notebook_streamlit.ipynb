```python
import pandas as pd
import seaborn as sns
import matplotlib as plt


link ="https://raw.githubusercontent.com/murpi/wilddata/master/quests/cars.csv"
df_cars = pd.read_csv(link)
df_cars['year']=df_cars['year'].replace(',', '', regex=False)
df_cars['year']= df_cars['year'].astype(int)
```

```python
df_cars.info()
```

```python
df_cars['year']=pd.to_datetime(df_cars['year'], format='%Y',errors='coerce')
df_cars['year']=df_cars['year'].dt.year
print(df_cars['year'])
df_cars
```

```python
df_cars.info()
```

```python
df_cars['continent']=df_cars['continent'].str.strip().replace('.', '', regex=False)
df_cars
```

```python
df_cars['continent'] = df_cars['continent'].str.rstrip()
df_cars['continent'] = df_cars['continent'].replace('.', '', regex=False)
df_cars
```

```python
df_cars.describe()
```

```python
df_cars.drop_duplicates()
```

```python
df_cars.to_csv('new_df_cars.csv', index=False)
```
